### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
data_1['HP to WT Ratio'] = data_1['horsepower']/data_1['weight']
best_ratio = round(data_1['HP to WT Ratio'].max(),4)
median = data_1['mpg'].median()
print(f"Best horsepower to weight ratio are : {best_ratio}")

data_1[(data_1['mpg']>median)].sort_values('HP to WT Ratio',ascending=False).head(3)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
plt.figure(figsize=(10,6))
sns.lineplot(data_1[data_1['origin']=='japan'], x='model_year',y='mpg',color = 'red',label='Japan')
sns.lineplot(data_1[data_1['origin']=='usa'], x='model_year',y='mpg',color = 'blue',label='USA')
sns.lineplot(data_1[data_1['origin']=='europe'], x='model_year',y='mpg',color = 'green',label='Europe')
plt.grid(alpha=0.30)
plt.legend()
plt.tight_layout()
plt.title("Evolution of MPG over the years (origin specific)")
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
task_01 = data_1.groupby('name').agg(Count = ('model_year','count')).sort_values('Count',ascending=False).reset_index()
task_01[task_01['Count']>1]
```

```python
names = task_01[task_01['Count']>1]['name'].to_list()
task_02 = pd.DataFrame()
name = []
std_ = []
for i in names:
    df = data_1.loc[:,['name','model_year','mpg']]
    df = df[df['name']==i]
    name.append(i)
    std = df['mpg'].std()
    std_.append(std)
task_02['name'] = name
task_02['std mpg'] = std_
task_02
```

```python
task_03 = task_02[task_02['std mpg']<1.0].reset_index()
task_03
```

```python
task_04 = data_1.groupby(['name'])['model_year'].count().rename('No. of appearances').sort_values(ascending=False).reset_index()
names_04 = task_04['name'].to_list()
avg_mpg = []
for i in names_04:
     mpg = data_1[data_1['name']==i]['mpg'].mean()
     avg_mpg.append(mpg)
task_04['Avg MPG'] = avg_mpg
task_04
```
